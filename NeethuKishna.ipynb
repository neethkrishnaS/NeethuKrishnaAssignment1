```python
df['mean_Score']=((df['math score']+df['reading score']+df['writing score'])/3).round()
df.head(37)
```

```python
# a.Find out how many males and females participated in the test.
df['gender'].value_counts()
```

```python
#label encoding as we want it in integer form 
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
df['gender']=lc.fit_transform(df['gender'])
df['race/ethnicity']=lc.fit_transform(df['race/ethnicity'])
df['parental level of education']=lc.fit_transform(df['parental level of education'])
df['lunch']=lc.fit_transform(df['lunch'])
df['test preparation course']=lc.fit_transform(df['test preparation course'])
df.head(50)
```

```python
#What do you think about the students' parental level of education
df['parental level of education'].value_counts()
```

```python
sns.barplot(x='parental level of education',y='mean_Score',data=df)
```

```python
#so from plot we can say students whose parental level of education is MASTERS DEGREE has highest mean score becasue masters degree belong to 3rd category
#and students whose parental level of education is HIGH SCHOOL has lowest mean score because high school belong to 2nd category.
```

```python
# d.What do you think about the scoring variation for math, reading and writing based on
# ii.Test preparation course
from scipy.stats import norm
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Gender vs Scores for all three subjects:', fontsize=20)
sns.boxplot(x='gender', y='math score', data=df, ax=ax1)
ax1.title.set_text('Math score vs Gender')
sns.boxplot(x='gender', y='reading score', data=df, ax=ax2)
ax2.title.set_text('Reading score vs Gender')
sns.boxplot(x='gender', y='writing score', data=df, ax=ax3)
ax3.title.set_text('Writing score vs Gender')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Test Preparation Course vs Scores for all three subjects:', fontsize=20)
sns.boxplot(x='test preparation course', y='math score', data=df, ax=ax1)
ax1.title.set_text('Math score vs Test Preparation Course')
sns.boxplot(x='test preparation course', y='reading score', data=df, ax=ax2)
ax2.title.set_text('Reading score vs Test Preparation Course')
sns.boxplot(x='test preparation course', y='writing score', data=df, ax=ax3)
ax3.title.set_text('Writing score vs Test Preparation Course')
```
