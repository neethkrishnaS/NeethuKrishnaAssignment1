```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
datset = pd.read_csv("titanic_dataset.csv")
datset['Cabin']
```

```python
print(np.where(datset['Pclass']>0.12))
```

```python
from scipy import stats
 
z = np.abs(stats.zscore(datset['Pclass']))
print(z)
```

```python
Q1 = np.percentile(datset['Pclass'], 25, method='midpoint')
Q3 = np.percentile(datset['Pclass'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
upper = Q3+1.5*IQR
upper_array = np.array(datset['Pclass'] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())
lower = Q1-1.5*IQR
lower_array = np.array(datset['Pclass'] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
 #Do min max scaling on the feature set
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
datset = pd.read_csv("titanic_dataset.csv")
X=datset[["PassengerId","Pclass","Age"]]
Y=datset[["Survived"]]
X_train,X_test,y_train,y_test = train_test_split(X,Y,train_size=0.7,test_size=0.3,random_state=51)
norm_mmc=MinMaxScaler()
norm_mmc_trans_X_train=norm_mmc.fit_transform(X_train)
norm_mmc_trans_X_test=norm_mmc.fit_transform(X_test)
norm_mmc_trans_X_train_df=pd.DataFrame(norm_mmc_trans_X_train)
norm_mmc_trans_X_train_df.head()
```

```python
norm_mmc_trans_X_train_df=pd.DataFrame(norm_mmc_trans_X_train,columns=["PassengerId","Pclass","Age"])
norm_mmc_trans_X_train_df.head()
```

```python
norm_mmc_trans_X_test_df=pd.DataFrame(norm_mmc_trans_X_test,columns=["PassengerId","Pclass","Age"])
norm_mmc_trans_X_test_df.head()
```
